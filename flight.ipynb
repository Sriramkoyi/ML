```python
import pandas as pd

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
rbf=SVC(kernel='rbf')
```

```python
rbf.fit(X_train,y_train)
```

```python
y_pred1=rbf.predict(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
poly=SVC(kernel='poly')
poly.fit(X_train,y_train)
y_pred11=poly.predict(X_test)
print(classification_report(y_test,y_pred11))
print(confusion_matrix(y_test,y_pred11))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
y_pred110=sigmoid.predict(X_test)
print(classification_report(y_test,y_pred110))
print(confusion_matrix(y_test,y_pred110))
```

```python
from sklearn.model_selection import GridSearchCV

param={'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['rbf']}

```

```python
grid=GridSearchCV(SVC(),param_grid=param,refit=True,cv=5,verbose=3)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred110=grid.predict(X_test)
print(classification_report(y_test,y_pred110))
print(confusion_matrix(y_test,y_pred110))
```
